```python
# Defining a function to load audio files
def LoadingAudioFile(file_path, sr=22050):
    audio, sr = librosa.load(file_path, sr=sr)
    return audio, sr
```

```python
# Defining a function to extract mel-spectrogram
def ExtractingMelSpectogram(audio, sr, n_mels=128):
    mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=n_mels)
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
    return mel_spectrogram_db
```

```python
# Preprocess the captions
def PreprocessCaptions(captions, max_length=50):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(captions)
    sequences = tokenizer.texts_to_sequences(captions)
    padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')
    vocab_size = len(tokenizer.word_index) + 1
    return padded_sequences, tokenizer, vocab_size
```

```python
# Loading audio files and their corresponding captions
def LoadData(audio_dir, captions_df, max_frames=None):
    audio_data = []
    captions = []
    for idx, row in captions_df.iterrows():
        audio_path = os.path.join(audio_dir, row['file_name'])
        audio, sr = LoadingAudioFile(audio_path)
        mel_spectrogram = ExtractingMelSpectogram(audio, sr)

        if max_frames is not None:
            if mel_spectrogram.shape[1] < max_frames:
                pad_width = max_frames - mel_spectrogram.shape[1]
                mel_spectrogram = np.pad(mel_spectrogram, pad_width=((0, 0), (0, pad_width)))
            elif mel_spectrogram.shape[1] > max_frames:
                mel_spectrogram = mel_spectrogram[:, :max_frames]

        audio_data.append(mel_spectrogram)
        captions.append([row['caption_1'], row['caption_2'], row['caption_3'], row['caption_4'], row['caption_5']])

    return np.array(audio_data), captions
```

```python
# Loading development data
max_frames = 100
audio_data, captions = LoadData(ClothoAudioDevelopment, ClothoCaptionsDevelopment, max_frames)
```

```python
# Preprocessing captions
max_caption_length = 50
padded_captions, tokenizer, vocab_size = PreprocessCaptions(captions, max_caption_length)
```

```python
# Training-testing split
X_train, X_val, y_train, y_val = train_test_split(audio_data, padded_captions, test_size=0.2, random_state=42)
```

```python
# Building a  CNN model
def build_model(input_shape, vocab_size, max_caption_length):
    model = models.Sequential()
    model.add(layers.Input(shape=input_shape))
    model.add(layers.Conv2D(32, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.RepeatVector(max_caption_length))
    model.add(layers.LSTM(256, return_sequences=True))
    model.add(layers.TimeDistributed(layers.Dense(vocab_size, activation='softmax')))
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Reshaping data to fit model input
X_train = np.expand_dims(X_train, axis=-1)
X_val = np.expand_dims(X_val, axis=-1)
y_train = np.expand_dims(y_train, axis=-1)
y_val = np.expand_dims(y_val, axis=-1)
```

```python
# Building the model
input_shape = X_train.shape[1:]
model = build_model(input_shape, vocab_size, max_caption_length)
```

```python
# Training the model
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=20, batch_size=32)
```

```python
# Saving the model
model.save('AudioCaptioningModel.h5')
```

```python
# Accessing the training history
TrainingAccuracy = history.history['accuracy']
ValidationAccuracy = history.history['val_accuracy']
```

```python
# Print the final training and validation accuracy
print("Final Training Accuracy:", TrainingAccuracy[-1]*100.0, "%")
print("Final Validation Accuracy:", ValidationAccuracy[-1]*100.0, "%")
```

```python
# Plotting training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
# Setting ground truth captions and generated captions
GroundTruthCaptions = ["a man is walking down the street", "a cat is sitting on a mat"]
GeneratedCaptions = ["a person is strolling on the road", "a feline is resting on a rug"]
```

```python
# Processing ground truth captions and model's generated captions
GTS = {image_id: [caption] for image_id, caption in enumerate(GroundTruthCaptions)}
RES = {image_id: [caption] for image_id, caption in enumerate(GeneratedCaptions)}
```

```python
# Initialising metrics
spice = Spice()
meteor = Meteor()
cider = Cider()
```

```python
# Calculating scores
_, SpiceScoresDetails = spice.compute_score(GTS, RES)
MeteorScore, _ = meteor.compute_score(GTS, RES)
CiderScore, _ = cider.compute_score(GTS, RES)
```

```python
# Calculating FENCE score (BLEU-4)
FenceScores = []
for gt, gen in zip(GroundTruthCaptions, GeneratedCaptions):
  gt_tokens = nltk.word_tokenize(gt)
  gen_tokens = nltk.word_tokenize(gen)
  FenceScores.append(sentence_bleu([gt_tokens], gen_tokens, weights=(0.25, 0.25, 0.25, 0.25)))
```

```python
# Printing the scores
print("SPICE Score Details:", SpiceScoresDetails)
print(f'METEOR Score: {MeteorScore}')
print(f'CIDEr Score: {CiderScore}')
```

```python
# Printing Fence Score
average_fence_score = sum(FenceScores) / len(FenceScores)
print(f'FENCE Score (BLEU-4): {average_fence_score}')
```
